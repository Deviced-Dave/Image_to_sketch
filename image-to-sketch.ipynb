```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
img = cv2.imread("sample_images/Joseph_Engelberger.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,8))
plt.imshow(img_invert, cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21), sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing, cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final, cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

```python
def sketchify(path):
    # Read the image
    # Change color code from BGR to RGB
    
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Convert image to GrayScale for easy handling
    
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    # Invert image colors for detailed study
    
    img_invert = cv2.bitwise_not(img_gray)
    
    # Smoothing the image
    
    img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
    
    # Converting to Sketch
    
    final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
    
    return img, img_gray, img_invert, img_smoothing, final
```

```python
def sketchify_display(img, img_gray, img_invert, img_smoothing, final):
    plt.figure(figsize=(20,20))
    plt.subplot(1,5,1)
    plt.imshow(img)
    plt.axis("off")
    plt.title("Original Image")
    plt.subplot(1,5,2)
    plt.imshow(img_gray,cmap="gray")
    plt.axis("off")
    plt.title("GrayScale Image")
    plt.subplot(1,5,3)
    plt.imshow(img_invert,cmap="gray")
    plt.axis("off")
    plt.title("Inverted Image")
    plt.subplot(1,5,4)
    plt.imshow(img_smoothing,cmap="gray")
    plt.axis("off")
    plt.title("Smoothen Image")
    plt.subplot(1,5,5)
    plt.imshow(final,cmap="gray")
    plt.axis("off")
    plt.title("Final Sketch Image")
    plt.show()
```

```python
path = "sample_images/Ray_Bradbury.jpg"
img, img_gray, img_invert, img_smoothing, final = sketchify(path)
sketchify_display(img, img_gray, img_invert, img_smoothing, final)
```
